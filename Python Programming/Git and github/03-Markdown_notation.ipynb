## What is Markdown 

Markdown is a lightweight markup language used to format text using simple, easy-to-read syntax. It allows users to create formatted documents with elements like headers, lists, links, images, and code blocks without the complexity of HTML. Markdown is widely used in websites, documentation, and readme files due to its simplicity and portability. It can be converted into various formats, such as HTML or PDF, making it versatile for different platforms. Its syntax is designed to be readable as plain text, even before being rendered into its final formatted form.


## Write headlines

Nothing could be simpler, just put a _#_ to design that your text is a title. For example :


```markdown
# This is a title
```


The more #'s you add, the lower you go in the hierarchy of importance of the titles. For example:


```markdown
## This is a subtitle
### This is a sub-subtitle
#### This is a sub-sub-subtitle
##### etc
#### This is the maximum
```


### Put a link

To put a link in Github, you have to proceed as follows:


```markdown
[The text to link](the url to redirect to)
```

For example


```markdown
[Jedha](https://jedha.co)
```
