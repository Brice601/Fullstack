```python
df = pd.read_csv("Speed+Dating+Data.csv", encoding='windows-1252')  # dérivé de encoding='latin-1'
df
```

```python
df.info()
```

```python
for i, col in enumerate(df.columns):
    print(f"{i}: {col}")
```

```python
# ajout colonne label genre et race

df["gender_label"] = df["gender"].map({0: "Femme", 1: "Homme"})
df["race_label"] = df["race"].map({1: "Noir", 2: "Blanc", 3:"Hispanique/Latino", 4: "Asiatique", 5: "Amérindien", 6: "Autre"})
```

```python
# visualisation nb Hommes Femmes par race

fig_gender_race = px.histogram(df, x="race_label", color="gender_label")

fig_gender_race.update_layout(xaxis_title='Race', yaxis_title="Nombre")
fig_gender_race.update_layout(legend_title_text='Genre')
fig_gender_race.show()
```
